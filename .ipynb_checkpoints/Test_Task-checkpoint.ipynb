# Анализ отзывов на банки

В данных представлены отзывы на банки с сайта [banki.ru](https://www.banki.ru/)

- title - название отзыва
- rating - оценка по отзыву, варьируется от 1 до 5
- bank_name - название банка
- date - дата публикации отзыва

Необходимо EDA (Exploratory Data Analysis) по этим данным и сделать несколько выводов (приветствуются визуализации):
- назвать топ-15 банков по числу отзывов
- какой банк из числа топ-15 по отзывов имеет самый высокий средний рейтинг?
- сделать выводы об изменении среднего рейтинга банков по месяцам из топ-5 банков по числу отзывов
- нет ли в данных в целом каких-то странностей на ваш взгляд?
- дополнительный EDA на ваше усмотрение, возможные интересные наблюдения

Необходимо выбрать метрику и построить бинарный классификатор отзывов, добившись при этом лучшей метрики на валидационной выборке (отзывы >=3 считать положительными, <3 - отрицательными)

```python
# Определяем топ-15 банков по числу отзывов
top_banks = df['bank_name'].value_counts().head(15)

# Визуализация топ-15 банков по числу отзывов
plt.figure(figsize=(12, 6))
top_banks.plot(kind='bar', color='skyblue')
plt.title('Топ-15 банков по числу отзывов')
plt.xlabel('Название банка')
plt.ylabel('Число отзывов')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
bank_reviews_stats = df.groupby('bank_name').agg({'rating': ['count', 'mean']})
bank_reviews_stats.columns = ['count_reviews', 'average_rating']

# Отфильтруем только топ-15 банков
top_15_banks_stats = bank_reviews_stats.loc[top_banks.index]

# Найдем банк с самым высоким средним рейтингом среди топ-15 банков
highest_avg_rating_bank = top_15_banks_stats['average_rating'].idxmax()
highest_avg_rating = top_15_banks_stats.loc[highest_avg_rating_bank, 'average_rating']

print(f"Банк с самым высоким средним рейтингом среди топ-15 банков: {highest_avg_rating_bank} (средний рейтинг: {highest_avg_rating:.2f})")
```

```python
# Сначала сгруппируем данные по банку и рейтингу
bank_rating_grouped = df.groupby(['bank_name', 'rating']).size().unstack()

# Найдем банк с самым высоким средним рейтингом
highest_avg_rating_bank = top_15_banks_stats['average_rating'].idxmax()

# Выберем данные только для банка с самым высоким рейтингом
highest_rating_bank_data = bank_rating_grouped.loc[highest_avg_rating_bank]

# Создадим график частоты отзывов для банка с самым высоким рейтингом
plt.figure(figsize=(10, 6))
plt.bar(highest_rating_bank_data.index, highest_rating_bank_data.values, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Частота отзывов для банка "{highest_avg_rating_bank}" по рейтингам')
plt.xlabel('Рейтинг')
plt.ylabel('Частота отзывов')
plt.xticks(highest_rating_bank_data.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 1. Выбираем топ-5 банков по числу отзывов
top_5_banks = df['bank_name'].value_counts().head(5).index

# 2. Преобразуем дату публикации отзыва в формат даты и извлечем месяц
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month

# 3. Группируем данные по месяцам и банкам
grouped_data = df[df['bank_name'].isin(top_5_banks)].groupby(['bank_name', 'month'])

# 4. Вычисляем средний рейтинг для каждого месяца и каждого банка
average_rating_by_month = grouped_data['rating'].mean().unstack(level=0)

# 5. Визуализируем изменение среднего рейтинга для каждого банка по месяцам
plt.figure(figsize=(12, 6))
for bank in average_rating_by_month.columns:
    plt.plot(average_rating_by_month.index, average_rating_by_month[bank], marker='o', label=bank)

plt.title('Изменение среднего рейтинга по месяцам (Топ-5 банков)')
plt.xlabel('Месяц')
plt.ylabel('Средний рейтинг')
plt.xticks(range(1, 13), ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек'])
plt.legend()
plt.grid(True)
plt.show()
```

```python
import nltk
import spacy
nltk.download('stopwords')
nltk.download('wordnet')
nlp = spacy.load('ru_core_news_lg')
```

```python
import re
import string
import html
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

def clean_string(text, stem="None"):

    final_string = ""

    # Make lower
    text = text.lower()

    # Remove line breaks
    text = re.sub(r'\n', '', text)

    text = re.sub(r'(@.*?)[\s]', ' ', text)
    text = re.sub(r'<br />', '', text)

    # Decode HTML entities
    text = html.unescape(text)

    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)

    # Remove trailing whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove puncuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Remove stop words
    text = text.split()
    useless_words = nltk.corpus.stopwords.words("russian")

    text_filtered = [word for word in text if not word in useless_words]

    # Remove numbers
    text_filtered = [re.sub(r'\w*\d\w*', '', w) for w in text_filtered]

    # Stem or Lemmatize
    if stem == 'Stem':
        stemmer = PorterStemmer()
        text_stemmed = [stemmer.stem(y) for y in text_filtered]
    elif stem == 'Lem':
        lem = WordNetLemmatizer()
        text_stemmed = [lem.lemmatize(y) for y in text_filtered]
    elif stem == 'Spacy':
        text_filtered = nlp(' '.join(text_filtered))
        text_stemmed = [y.lemma_ for y in text_filtered]
    else:
        text_stemmed = text_filtered

    final_string = ' '.join(text_stemmed)

    return final_string
```

```python
print('Original: ', df.body[0])
print('Processed: ', clean_string(df.body[0]))
```
